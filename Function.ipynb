```python
# 5 What are iterators in Python and how do they differ from iterables?
""" In Python, an iterator is an object that implements the iterator protocol, which consists of the methods __iter__() and __next__().
An iterator can be used to iterate over a sequence of values, one at a time, using the next() function. Iterators are often used to iterate over large datasets or to create custom iteration protocols.
Iterables, on the other hand, are objects that can be iterated over, but they do not necessarily implement the iterator protocol. Examples of iterables include lists, tuples, and dictionaries. An iterable can be converted to an iterator using the iter() function, which returns an iterator object that can be used to iterate over the values in the iterable. In summary, all iterators are iterables, but not all iterables are iterators. """
```

```python
# 6  Explain the concept of generators in Python and how they are defined.
""" Generators in Python are a special type of iterator that allow you to create iterators using a simple and concise syntax.
Generators are defined using the yield statement, which allows you to return a value from the generator function without losing its state.
When a generator function is called, it returns a generator object that can be used to iterate over the values produced by the generator. Each time the next() function is called on the generator object, the generator function resumes execution from where it left off and continues until it hits the next yield statement or returns.
This makes generators a powerful tool for creating iterators that can produce large sequences of values without consuming a lot of memory. """
```

```python
# 7. What are the advantages of using generators over regular functions?
""" The advantages of using generators over regular functions include:
1. Memory efficiency: Generators produce values on the fly and do not store the entire sequence in memory, making them more memory-efficient for large datasets.
2. Lazy evaluation: Generators only compute values when they are needed, which can improve performance and reduce computation time for large datasets.
3. Simplicity: Generators can be defined using a simple and concise syntax, making them easier to read and understand.
4. Infinite sequences: Generators can produce infinite sequences of values, which is not possible with regular functions that return a fixed number of values.
5. State retention: Generators retain their state between calls, allowing them to produce values in a more flexible and dynamic way than regular functions."""
```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
""" The map(), reduce(), and filter() functions in Python are all higher-order functions that operate on iterables, but they have different purposes and behaviors:
1. map(): The map() function applies a given function to each item in an iterable and returns an iterator that produces the results. It is used for transforming the items in the iterable by applying the specified function to each item.
2. reduce(): The reduce() function applies a given function of two arguments cumulatively to the items in an iterable, reducing the iterable to a single value. It is used for performing a rolling computation on the items in the iterable, such as summing or multiplying them together. The reduce() function is not built into Python by default and must be imported from the functools module.
3. filter(): The filter() function constructs an iterator from elements of an iterable for which a function returns true. It is used for filtering the items in the iterable based on a specified condition, returning only those items that satisfy the condition.
The filter() function takes two arguments: the first argument is the function that defines the filtering condition, and the second argument is the iterable to be processed.
The map() and filter() functions return iterators, while reduce() returns a single value. Additionally, map() and filter() can be used with any callable object, while reduce() requires a function that takes two arguments."""
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.?
class FibonacciIterator:
    """
    An iterator class that generates the Fibonacci sequence up to a specified number of terms.
    """

    def __init__(self, n):
        """
        Initializes the Fibonacci iterator with the number of terms.

        Args:
            n (int): The number of terms in the Fibonacci sequence.
        """
        self.n = n
        self.a, self.b = 0, 1
        self.current_term = 0

    def __iter__(self):
        """
        Returns the iterator object itself.
        """
        return self

    def __next__(self):
        """
        Returns the next term in the Fibonacci sequence.

        Raises:
            StopIteration: If the number of terms has been reached.
        """
        if self.current_term >= self.n:
            raise StopIteration

        result = self.a
        self.a, self.b = self.b, self.a + self.b
        self.current_term += 1
        return result


# Example usage:
if __name__ == "__main__":
    num_terms = 10
    fibonacci_iterator = FibonacciIterator(num_terms)

    print(f"Fibonacci sequence up to {num_terms} terms:")
    for term in fibonacci_iterator:
        print(term)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
for line in read_file_line_by_line('example.txt'):
    print(line)

```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Example list of tuples
tuples_list = [(1, 3), (4, 1), (2, 2), (5, 0)]

# Sort the list of tuples based on the second element
sorted_list = sorted(tuples_list, key=lambda x: x[1])

print(sorted_list)  # Output: [(5, 0), (4, 1), (2, 2), (1, 3)]
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40]

# Convert Celsius to Fahrenheit using map() and a lambda function
fahrenheit_temperatures = list(map(lambda c: (c * 9/5) + 32, celsius_temperatures))

print(fahrenheit_temperatures)  # Output: [32.0, 68.0, 86.0, 104.0]
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

# Input string
input_string = "This is an example string."

# Use filter() to remove vowels
filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)  # Output: "Ths s n xmpl strng."
```
